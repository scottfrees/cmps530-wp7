```python
import matplotlib.colors as colors
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from sklearn.decomposition import PCA
from sklearn.datasets import load_digits
digits = load_digits()
digits.data.shape
```

```python
pca = PCA(2)  # project from 64 to 2 dimensions
projected = pca.fit_transform(digits.data)
print(digits.data.shape)
print(projected.shape)
```

```python
plt.scatter(projected[:, 0], projected[:, 1],
            c=digits.target, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('Spectral', 10))
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar();
```

```python
digits.data[0]
```

```python
import random
sample_index = 1007#random.randint(0, len(digits.target))
target = digits.target[sample_index]
sample = pca.transform([digits.data[sample_index]])

norm = colors.Normalize(vmin=0, vmax=10)
cmap = plt.get_cmap('Spectral', 10)
plt.scatter(projected[:, 0], projected[:, 1],
            c=digits.target, edgecolor='none', alpha=0.5,
            cmap=cmap)
plt.scatter([sample[0][0]], [sample[0][1]],s = 100, c='magenta', marker=str('x'), cmap=cmap)
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.title(f'Sample #{sample_index} = {target}')
plt.colorbar(plt.cm.ScalarMappable(norm=norm, cmap=cmap));
```
